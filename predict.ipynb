```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import pandas as pd
import os
import tensorflow as tf
# from tensorflow.keras.preprocessing.image import ImageDataGenerator
# from tensorflow.keras.models import load_model

# suppress logs
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

# python predict.py -d "C:\Users\Dell\PycharmProjects\pythonProject\dataset\DATA\test" -m "C:\Users\Dell\PycharmProjects\pythonProject\models\model_vgg_08_08_2023_12_00.hdf5" -i (150, 150, 3)
# ap = argparse.ArgumentParser()
# ap.add_argument('-d', '--dataset', required=True, help='type of images: [train, valid, test]')
# ap.add_argument('-m', '--model', required=False, help='path to model')
# ap.add_argument('-i', '--input_shape', required=True, help='Requirement image input shape by model. e. g. (150, 150, 3) for VGG, (224, 224, 3) for ResNet')
# args = vars(ap.parse_args())
args ={
    'dataset': '/content/drive/MyDrive/Colab Notebooks/traffic_sign_classifier/data/test',
    'model': '/content/drive/MyDrive/Colab Notebooks/traffic_sign_classifier/output_model_rn/model_09_08_2023_08_10.hdf5',
}

datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1. / 255.
)

generator = datagen.flow_from_directory(
    directory=args['dataset'],
    target_size=(224, 224),
    batch_size=1,
    class_mode='sparse',
    shuffle=False
)

model = tf.keras.models.load_model(args['model'])

y_prob = model.predict_generator(generator, workers=1)

y_true = generator.classes

print(y_prob)
```

```python
y_prob.shape
```

```python
labels = pd.read_csv('/content/labels.csv')
labels
```

```python
import numpy as np

column_names = labels['ClassId'].tolist()
index_names = generator.filenames

# Tworzenie DataFrame z tablicy NumPy i określenie nazw kolumn i indeksów
df = pd.DataFrame(y_prob, columns=column_names, index=index_names)
df.head(3)
```

```python
y_true = generator.classes
y_pred = df.idxmax(axis=1)

predictions = pd.DataFrame({'y_pred': y_pred, 'y_true': y_true}, index=generator.filenames)

def check_errors(row):
    if row['y_pred'] == row['y_true']:
        return 0
    else:
        return 1

predictions['errors'] = predictions.apply(check_errors, axis=1)

predictions.head(3)
```

```python
# print(f"[INFO] Macierz konfuzji:\n{confusion_matrix(predictions['y_true'].values, predictions['y_pred'].values)}")
print(f"[INFO] Raport klasyfikacji:\n{classification_report(predictions['y_true'].values, predictions['y_pred'].values)}")
print(f"[INFO] Dokładność modelu: {accuracy_score(predictions['y_true'].values, predictions['y_pred'].values) * 100:.2f}%")
```
